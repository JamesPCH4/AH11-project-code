```python
import pandas as pd
```

```python
df = pd.DataFrame()
df1 = pd.read_json("/content/alignn/alignn/form_f11/history_train.json")
df2 = pd.read_json("/content/alignn/alignn/form_f11/history_val.json")

df["Training"]=df1.iloc[:,0]
df["validation"]=df2.iloc[:,0]

list = []
calc = 1

while int(calc) < 301:
    list.append(calc)
    calc = int(calc) + 1
df["Epoch"]=list

import matplotlib.pyplot as plt

df.plot(x="Epoch", y=["Training", "validation"],ylabel="Mean absolute error",title="Training and validation loss against epoch for formation energy per atom")
plt.show()
```

```python
df_form = pd.read_csv("/content/alignn/alignn/form_f11/prediction_results_test_set.csv")

df_form.plot(x="target",y="prediction",kind="scatter",xlabel="Target",ylabel="Prediction",title="Prediction against target for formation energy per atom")
plt.show()

from sklearn.metrics import r2_score

print("The r2 value is:",r2_score(df_form["target"], df_form["prediction"]))
```
